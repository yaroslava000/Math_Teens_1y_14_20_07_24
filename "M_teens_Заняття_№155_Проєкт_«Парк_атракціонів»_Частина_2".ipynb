## Варіант 1 Завдання 1.  Фудкорт «Морепродукти»

Передня частина вивіски має форму рівнобічної трапеції. Знайдіть площу цієї частини, якщо проекція бічної сторони на більшу основу — 70 см, кут між бічною стороною і основою — 60°, решта довжини основи трапеції — 6 м.

image.png

```python
import math

def s_of_bilbord(a, b, ang):
  whole_a = a+b
  whole_b = b-a
  c = math.tan(ang/180*math.pi)*a
  s = (whole_a+whole_b)/2*c
  return f"площа вивіски = {format(s, ".2f")} м^2"

print(s_of_bilbord(0.7, 6, 60))
```

Для цього використовуємо **set_xticks**, **set_yticks**:

Додатково можна створити тінь для кожної позначки атракціона.

Для цього створюємо змінну **shadow_effect**, значення якої використаємо для параметра **path_effects** в об'єктах позначок:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle, Polygon, RegularPolygon, FancyBboxPatch
import matplotlib.patheffects as path_effects

fig, ax = plt.subplots()
ax.set_xlim(-17, 17)
ax.set_ylim(-17, 19)
ax.set_facecolor("#557249")

shadow_effect = [path_effects.SimpleLineShadow(), path_effects.Normal()]

n_sides = 5
park_radius = 9
center = np.array([0, 0])
angles = np.linspace(0, 2*np.pi, n_sides + 1)
vertices = [park_radius*np.array([np.cos(a), np.sin(a)]) for a in angles]

park_road_1 = Polygon(vertices, fill=None, edgecolor="#818f65", linewidth=10, path_effects = shadow_effect)
park_road_2 = Polygon(vertices, fill=None, edgecolor="#aec198", linestyle="--", linewidth=2)
ax.add_patch(park_road_1)
ax.add_patch(park_road_2)

make_food_place = FancyBboxPatch(center-0.6, 1, 1, color="#e3e2cd", label="холодні напої", path_effects = shadow_effect)
ax.add_patch(make_food_place)

make_circle = Circle(vertices[0], 1.5, color="#e4845d", label="башта падіння", path_effects = shadow_effect)
ax.add_patch(make_circle)

place_star=[]
for i in range(10):
  angle = (np.pi/5)*i
  r = 2 if i % 2 == 0 else 1
  place_star.append(r*np.array([np.cos(angle), np.sin(angle)]))
make_star = Polygon(vertices[1]+place_star, color="#de5d47", label="американські гірки", path_effects = shadow_effect)
ax.add_patch(make_star)

place_diamond = [np.array([0, 2]), np.array([1.5, 0]), np.array([0, -2]), np.array([-1.5, 0])]
make_diamond = Polygon(vertices[2]+place_diamond, color = "#d53b2a", label = "лабіринт", path_effects = shadow_effect)
ax.add_patch(make_diamond)

make_square = Rectangle(vertices[3]-1, 2.5, 2.5, color = "#bb2c32", label = "маятник", path_effects = shadow_effect)
ax.add_patch(make_square)

angles_triangle = np.linspace(0, 2*np.pi, 4)
place_triangle = vertices[4]+[2*np.array([np.cos(a), np.sin(a)]) for a in angles_triangle]
make_triangle = Polygon(place_triangle, color = "#ba1f20", label = "замок", path_effects = shadow_effect)
ax.add_patch(make_triangle)

place_candy = Circle((0, 0), 2.2, color="#59a296", zorder=0)
ax.add_patch(place_candy)

park_name = FancyBboxPatch([-15, 12], 30, 4, boxstyle = 'sawtooth', color = "#a25965", path_effects = shadow_effect)
ax.add_patch(park_name)

ax.annotate("welcome to the Silent Hill for children", (0,14), color = '#9dbbb9', fontsize = 12, ha = 'center', va = 'center', fontweight = 'bold', fontname = 'Liberation Sans', path_effects = shadow_effect)

ax.set_aspect('equal')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.legend(loc = "lower right", fontsize = 9, ncol = 2)

ax.set_xticks([])
ax.set_yticks([])

plt.show()
```
