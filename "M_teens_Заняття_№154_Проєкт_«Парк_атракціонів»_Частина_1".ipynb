# Атракціон №1

## Варіант 1 Завдання 1.  Атракціон «Башта падіння»

Башта падіння складається з вертикальної основи та крісел.

Власне башта у формі квадрата, сидіння — по колу.
Тобто башта — вписаний квадрат у коло.

Довжина сторони основи башти (у формі квадрата) — 3 м. Скільки сидіть можна вмістити по колу (1 ряд сидінь), якщо ширина 1 сидіння — 70 см?

Створіть функцію для розрахунку довжини кола знаючи довжину вписаного квадрата а також порахуйте кількість сидінь.

image.png

```python
import math

def n(a, b):
  d = a*math.sqrt(2)
  c = d*math.pi
  n = c/b
  return f"на атракціон можна прилаштувати {format(n, ".0f")} сидінь"

print(n(3, 0.7))
```

# Атракціон 3

## Варіант 1 Завдання 3. Атракціон «Лабіринт»

Атракціон лабіринту спроектований у формі правильного многокутника. Ззовні стіни лабіринту треба покрити дзеркалами.

Знайдіть довжини зовнішніх стін лабіринта, тобто периметр многокутника, якщо навколо нього можна описати коло з радіусом 19 м, а центральний кут многокутника — 30°.

Створіть функцію для розрахунку периметра многокутника на основі радіуса описаного кола та центрального кута.

image.png

```python
import math

def p(r, a):
  n = 360/a
  b = r*2*math.sin(15/180*math.pi)
  p = n*b
  return f"лабіринт спроєктований у формі {n}-тикутника периметром {format(p, ".0f")} метрів"

print(p(19, 30))
```

# Атракціон 4

## Варіант 2 Завдання 4. Атракціон «Маятник»

На маятнику є максимальний кут розкачування. Знайдіть градусну міру цього кута, якщо границі маятника можна описати векторами BA, BC з координатами точок A(6;3), B(2;7), C(-1;2).

Не треба створювати функцію, наведіть розрахунки та їхні результати.

image.png

```python
import math

def angle(a, b):
   scalar_val = a[0]*b[0]+a[1]*b[1]
   a_module = (a[0]**2+a[1]**2)**0.5
   b_module = (b[0]**2+b[1]**2)**0.5
   angle_cos = scalar_val / a_module / b_module
   angle_rad = math.acos(angle_cos)
   angle_deg = format(angle_rad * 180 / math.pi, ".0f")
   return f"Кут між векторами ({a[0]};{a[1]}) і ({b[0]};{b[1]}): {angle_deg}°."

print(angle([4, -4],[-3, -5]))
```

## Варіант 1. Атракціон «Замок»

В дитячому замку можна потрапити в лабіринт, батути та м'який басейн.

Форма замку — прямокутник з прямокутними баштами з трикутними дахами; ширина замка — 15 м, довжина — 20 м. Знайдіть довжину паркана у формі кола, поставленого навколо цього замка. Також зауважте, що має бути відстань між всіма кутами замка до кола 2 м.

image.png

```python
import math

def park_lenght(a, b):
  c = math.sqrt(a**2+b**2)
  d = c+4
  C = d*math.pi
  return f"довжина парку становить {format(C, ".0f")} метрів"

print(park_lenght(15, 20))
```
